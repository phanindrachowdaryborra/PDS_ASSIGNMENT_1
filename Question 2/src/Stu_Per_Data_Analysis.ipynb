```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Cleaned_Students_Performance_Data.csv')
```

```python
# Set the style of seaborn
sns.set(style='whitegrid')
```

# 1. Box Plot

#Distribution of Scores (Math, Reading, Writing) by Gender.

#Melt the data to long format.

```python
scores_melted = data.melt(id_vars='gender', value_vars=['math score', 'reading score', 'writing score'],
                           var_name='Subject', value_name='Score')

plt.figure(figsize=(12, 6))
sns.boxplot(data=scores_melted, x='Subject', y='Score', hue='gender', palette='Set3')
plt.title('Distribution of Scores by Gender')
plt.ylabel('Scores')
plt.xlabel('Subjects')
plt.legend(title='Gender')
plt.show()
```

# 2. Bar Chart

#Average Scores by Parental Level of Education

```python
avg_scores = data.groupby('parental level of education')[['math score', 'reading score', 'writing score']].mean().reset_index()
avg_scores = avg_scores.melt(id_vars='parental level of education', var_name='Subject', value_name='Average Score')

plt.figure(figsize=(10, 6))
sns.barplot(x='parental level of education', y='Average Score', hue='Subject', data=avg_scores, palette='pastel')
plt.title('Average Scores by Parental Level of Education')
plt.ylabel('Average Score')
plt.xlabel('Parental Level of Education')
plt.xticks(rotation=45)
plt.legend(title='Subject')
plt.show()
```

# 3. Point Plot

#Effect of Lunch Program on Scores

```python
lunch_scores = data.groupby('lunch')[['math score', 'reading score', 'writing score']].mean().reset_index()
lunch_scores = lunch_scores.melt(id_vars='lunch', var_name='Subject', value_name='Average Score')

plt.figure(figsize=(8, 5))
sns.pointplot(x='lunch', y='Average Score', hue='Subject', data=lunch_scores, palette='Set1', markers=['o', 'D', 's'])
plt.title('Average Scores by Lunch Program')
plt.ylabel('Average Score')
plt.xlabel('Lunch Program')
plt.show()
```

# 4. Heat Map

#Impact of Test Preparation Course Completion on Scores


```python
# 4. Impact of Test Preparation Course Completion on Scores (Heatmap)
prep_scores = data.groupby('test preparation course')[['math score', 'reading score', 'writing score']].mean().reset_index()
prep_scores = prep_scores.set_index('test preparation course')

plt.figure(figsize=(8, 5))
sns.heatmap(prep_scores, annot=True, cmap='viridis', fmt='.1f', linewidths=.5)
plt.title('Average Scores Based on Test Preparation Course Completion')
plt.ylabel('Test Preparation Course')
plt.xlabel('Subjects')
plt.show()
```

# 5. Scatter Plot

# Math vs. Reading Scores, Colored by Race/Ethnicity

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='math score', y='reading score', hue='race/ethnicity', style='gender', palette='deep')
plt.title('Scatter Plot of Math vs. Reading Scores')
plt.xlabel('Math Score')
plt.ylabel('Reading Score')
plt.legend(title='Race/Ethnicity and Gender', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
