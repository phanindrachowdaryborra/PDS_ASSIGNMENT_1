# 1. Loading the Data

```python
import pandas as pd

df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

Check for the data types of the columns

```python
df.info()
```


Check for missing values

```python
df.isnull().sum()
```

#2. Handling Missing Values

Drop rows with significant missing values (if any).

Fill missing values in scores (if any) with the column's mean or median

Verify again if any missing values remain

```python
df.dropna(thresh=5, inplace=True)


df['math score'].fillna(df['math score'].mean(), inplace=True)
df['reading score'].fillna(df['reading score'].mean(), inplace=True)
df['writing score'].fillna(df['writing score'].mean(), inplace=True)

df.isnull().sum()
```

# 3. Standardizing Categorical Data

Removing extra spaces.

Correcting inconsistent case usage (e.g., “Male” vs “male”)

```python
df['gender'] = df['gender'].str.strip().str.lower()
df['race/ethnicity'] = df['race/ethnicity'].str.strip().str.lower()
df['parental level of education'] = df['parental level of education'].str.strip().str.lower()
df['lunch'] = df['lunch'].str.strip().str.lower()
df['test preparation course'] = df['test preparation course'].str.strip().str.lower()

print(df['gender'].unique())
print(df['race/ethnicity'].unique())
print(df['parental level of education'].unique())
print(df['lunch'].unique())
print(df['test preparation course'].unique())
```
